```python
# Cell 4: Load CSV
csv_path = "newdatset7.csv"  # change if needed
df = pd.read_csv(csv_path)
print("Columns:\n", df.columns.tolist())
print("\nPreview:\n")
display(df.head(1))
```

```python
# Cell 5: Parse nested JSON fields into Python objects
def parse_json_cell(cell_value):
    try:
        return json.loads(cell_value)
    except Exception:
        return []

data = {
    "eng_terms": parse_json_cell(df.loc[0, "CampusData.EngineeringBlock.search_terms"]),
    "eng_desc": df.loc[0, "CampusData.EngineeringBlock.description"],
    "eng_rooms": parse_json_cell(df.loc[0, "CampusData.EngineeringBlock.rooms"]),

    "svh_terms": parse_json_cell(df.loc[0, "CampusData.SVH_Block.search_terms"]),
    "svh_desc": df.loc[0, "CampusData.SVH_Block.description"],
    "svh_rooms": parse_json_cell(df.loc[0, "CampusData.SVH_Block.rooms"]),

    "fb_terms": parse_json_cell(df.loc[0, "CampusData.FacultyBlock.search_terms"]),
    "fb_desc": df.loc[0, "CampusData.FacultyBlock.description"],
    "fb_rooms": parse_json_cell(df.loc[0, "CampusData.FacultyBlock.rooms"]),

    "gf_terms": parse_json_cell(df.loc[0, "CampusData.GeneralFacilities.search_terms"]),
    "gf_desc": df.loc[0, "CampusData.GeneralFacilities.description"],
    "gf_places": parse_json_cell(df.loc[0, "CampusData.GeneralFacilities.places"]),
}

print("✅ Parsed JSON fields.")
print("Engineering rooms:", len(data["eng_rooms"]))
print("SVH rooms:", len(data["svh_rooms"]))
print("Faculty rooms:", len(data["fb_rooms"]))
print("General facilities:", len(data["gf_places"]))
```

```python
# Cell 6: Build a global search index across blocks and facilities

# Map each term to a target entity type and key
search_index = []

def add_terms(terms, entity_type, key):
    for t in terms:
        search_index.append({"term": t.lower(), "type": entity_type, "key": key})

add_terms(data["eng_terms"], "block", "engineering")
add_terms(data["svh_terms"], "block", "svh")
add_terms(data["fb_terms"], "block", "faculty")
add_terms(data["gf_terms"], "category", "general_facilities")

# Also index facility keywords for deeper search (e.g., "dominos", "chai")
for place in data["gf_places"]:
    for kw in place.get("Keywords", []):
        search_index.append({"term": kw.lower(), "type": "facility", "key": place.get("PlaceID")})

print(f"✅ Search index built with {len(search_index)} entries.")
```

```python
# Cell 7: Fuzzy search across the index
def fuzzy_lookup(query, limit=5):
    candidates = [item["term"] for item in search_index]
    results = process.extract(query.lower(), candidates, scorer=fuzz.QRatio, limit=limit)
    # results -> list of tuples: (matched_term, score, index)
    enriched = []
    for match_term, score, idx in results:
        item = search_index[idx]
        enriched.append({
            "match_term": match_term,
            "score": score,
            "type": item["type"],
            "key": item["key"]
        })
    return enriched

# Quick test
print(fuzzy_lookup("tech blok"))
```

```python
# Cell 8: Helpers for voice reply and formatting
def speak(text, lang="en"):
    tts = gTTS(text=text, lang=lang)
    tts.save("reply.mp3")
    return ipd.Audio("reply.mp3", autoplay=True)

def summary_for_block(block_key):
    if block_key == "engineering":
        desc = data["eng_desc"]
        rooms = data["eng_rooms"]
    elif block_key == "svh":
        desc = data["svh_desc"]
        rooms = data["svh_rooms"]
    elif block_key == "faculty":
        desc = data["fb_desc"]
        rooms = data["fb_rooms"]
    else:
        return "I couldn't find that block."

    # Build a concise summary including floor distribution
    floors = {}
    for r in rooms:
        floor = r.get("Location", "Unknown")
        floors[floor] = floors.get(floor, 0) + 1
    floor_summary = "; ".join([f"{f}: {c}" for f, c in floors.items()])
    return f"{desc} Rooms distribution — {floor_summary}."

def list_facilities_by_category(category):
    items = [p for p in data["gf_places"] if p.get("Category", "").lower() == category.lower()]
    if not items:
        return f"No facilities found in category '{category}'."
    lines = []
    for p in items:
        lines.append(f"- {p.get('Name')} — {p.get('Location')} (Nearby: {p.get('Nearby')})")
    return f"{category} facilities:\n" + "\n".join(lines)

def facility_details(place_id):
    for p in data["gf_places"]:
        if p.get("PlaceID") == place_id:
            name = p.get("Name")
            loc = p.get("Location")
            nearby = p.get("Nearby")
            cat = p.get("Category")
            return f"{name} ({cat}). Location: {loc}. Nearby: {nearby}."
    return "Facility not found."
```

```python
# Cell 9: Simple conversation memory
session = {
    "last_block": None,
    "last_facility": None,
    "last_category": None
}

def handle_query(query):
    # Intent shortcuts (category filters)
    q = query.lower().strip()

    # Category intents
    categories = ["food", "sports", "lab", "hostel", "store", "academic", "landmark"]
    for cat in categories:
        if cat in q or f"{cat}s" in q:
            session["last_category"] = cat.capitalize()
            return list_facilities_by_category(session["last_category"])

    # Fuzzy search
    results = fuzzy_lookup(query, limit=5)
    if not results:
        return "I couldn't find anything matching your query."

    top = results[0]  # Best match
    if top["type"] == "block":
        session["last_block"] = top["key"]
        return summary_for_block(top["key"])
    elif top["type"] == "facility":
        session["last_facility"] = top["key"]
        return facility_details(top["key"])
    elif top["type"] == "category":
        # General facilities category keyword
        return "You can ask for categories like Food, Sports, Hostels, Labs, or name specific places like 'Dominos' or 'Chai Garam'."
    else:
        return "I matched something, but I'm not sure how to answer that."
```

```python
# Cell 10: Build interactive UI
input_box = widgets.Text(
    description="Ask:",
    placeholder="e.g., tech block, dominos, food outlets, faculty block"
)
ask_button = widgets.Button(description="Search")
voice_toggle = widgets.Checkbox(description="Voice reply", value=True)
out = widgets.Output(layout={"border": "1px solid #ccc"})

def on_ask_clicked(b):
    with out:
        clear_output()
        query = input_box.value.strip()
        if not query:
            print("Please type a query.")
            return
        answer = handle_query(query)
        print(answer)
        if voice_toggle.value:
            display(speak(answer))

ui = widgets.VBox([
    widgets.HBox([input_box, ask_button, voice_toggle]),
    out
])

ask_button.on_click(on_ask_clicked)
display(ui)
print("✅ UI ready. Type a query and press Search.")
```
